```python
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.io as sio

def plot(x, y, ax, xlabel, ylabel, title, label):
    ax.plot(x,y, label=label)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.grid(b=True)

ex10_bifurcation = sio.loadmat("data/ttk4150/ex10_bifurcation.mat")
ex10_age19 = sio.loadmat("data/ttk4150/ex10_age19.mat")
ex10_age75 = sio.loadmat("data/ttk4150/ex10_age75.mat")
```

## 1: Wave splitting

```python
def compute_waves(P, Q, Zc):
    pf = (Zc*Q + P)/2
    pb = (P - Zc*Q)/2
    Qf = (Q + P/Zc)/2
    Qb = (Q - P/Zc)/2
    
    return pf, pb, Qf, Qb
```

## 2: Reflection from bifurcation

```python
t, P, Q, Zc = ex10_bifurcation['t'], ex10_bifurcation['P'], ex10_bifurcation['Q'], ex10_bifurcation['Zc']

pf, pb, Qf, Qb = compute_waves(P, Q, Zc)
Gamma = pb/pf

fig, (ax1, ax2, ax3) = plt.subplots(3,1, sharex=True)
fig.tight_layout()

plot(t, P, ax1, "", "$p$", "Pressure", "")
plot(t, Q, ax2, "", "$Q$", "Flow", "")
plot(t, Gamma, ax3, "", "$\Gamma$", "Reflection", "")

plt.show()
```

```python
fig, ax = plt.subplots(1,1, sharey=True)
fig.set_figwidth(15)
# Age 19 computation and plot
t, P, Q, Zc = ex10_age19['t'], ex10_age19['P'], ex10_age19['Q'], ex10_age19['Zc']
pf, pb, Qf, Qb = compute_waves(P, Q, Zc)

plot(t, P, ax, "", "$p$", "", "$p$")
plot(t, pf, ax, "", "$p_f$", "", "$p_f$")
plot(t, pb, ax, "", "$p_b$", "", "$p_b$")
fig.legend()
plt.show()
```
